```python
df= titanic.drop('Age',axis=1)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
df['Sex']= le.fit_transform(df['Sex'])
```

```python
df['Embarked']=df['Embarked'].replace('C',1)
```

```python
df
```

```python
from sklearn.impute import KNNImputer
imp= KNNImputer(n_neighbors=5)
df= imp.fit_transform(df)
```

```python
df
```

```python
df1= pd.DataFrame(df)
```

```python
df1
```

```python
df1.columns = ['Survived',	'Pclass',	'Sex',	'SibSp',	'Parch',	'Fare',	'Embarked']
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
y= titanic['Age']
```

```python
df1['Age']=y
```

```python
df1
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
it=IterativeImputer()
```

```python
x=it.fit_transform(df1)
```

```python
df2= pd.DataFrame(x)
```

```python
df2
```

```python
df2.columns = ['Survived',	'Pclass',	'Sex',	'SibSp',	'Parch',	'Fare',	'Embarked','Age']
```

```python
df2
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.catplot(x='Sex',hue='Survived',kind='count',data=df2)
```

```python
sns.violinplot(x='Sex',y='Age',hue='Survived', data=df2,split=True)
```

```python
df2['Range']=pd.qcut(df2['Fare'],4)

sns.barplot(x='Range',y='Survived',data=df2)
```

```python
df2['family']= 0

df2['family']=df2['Parch']+ df2['SibSp']
```

```python
df2['family'].unique()
```

```python
sns.factorplot(x='family',y='Survived',data=df2)
```

```python
x= df2.iloc[:,1:7]
y= df2.iloc[:, 0]
```

```python
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
```

```python
y
```

```python
lr.fit(x,y)
```

```python
lr.score(x,y)
```

```python
from sklearn.naive_bayes import MultinomialNB
multiNB= MultinomialNB()
multiNB.fit(x,y)
pred=multiNB.predict(x)
```

```python
from sklearn.naive_bayes import GaussianNB
gussNB= GaussianNB()
gussNB.fit(x,y)
pred_g=gussNB.predict(x)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,pred_g)
```

```python
from sklearn.tree import DecisionTreeClassifier
dc= DecisionTreeClassifier()
dc.fit(x,y)
pred= dc.predict(x)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,pred)
```

```python
from sklearn.svm import SVC
svc= SVC()
svc.fit(x,y)
pred= svc.predict(x)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier()
knn.fit(x,y)
pred=knn.predict(x)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,pred)
```
